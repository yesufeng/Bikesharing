#Data Mining of a Bikeshare dataframe

The data was given as a data.csv file where each row is a record of one ride, columns are: 
    
    bikeid (every bike has an unique bikeid), 
    start and end station ids; 
    start time and end time. 

The dataframe contains 1802997 records over the timeframe of two months (between 09/26/2013 to 11/25/2013). No other information is given.

The current analysis first evaluates the overall statistics of the data such as: average trip time, average time a bike spent at a station, at least how many records are missing; then the daily and hourly usage of bike stations in different community groups are compared.

## Load the data

the datetime strings are converted to timestamps, the columns are reordered to facilitate intepretation, the records are also sorted by starttime to facilitate later treatment such as estimation of minimum number of missing records

```python
data = pd.read_csv('data.csv')
data['stoptime'] = pd.to_datetime(data['stoptime'],unit = 's')
data['starttime'] = pd.to_datetime(data['starttime'],unit = 's')

deltatime = (data['stoptime'] - data['starttime'])
data['deltatime'] = deltatime/np.timedelta64(1,'s')
data = data.ix[:,['bikeid','start station id','end station id','starttime','stoptime','deltatime']]
data.sort(columns = 'starttime',ascending = True, inplace = True)
```

## 2. Fraction of missing records

If you track each bike by the bikeid, you should be able to track down each and every station the bike has been to. If the record is disconnected (i.e., the next start station is not the current end station after sorting the bike's record by time, then you know at least one record is missing connecting these two stations)

```python
bikeid_gb = data.groupby('bikeid')

def count_missing(df):
    start_st = df['start station id'].values
    end_st = df['end station id'].values
    test = [x!=y for (x,y) in zip(start_st[1:],end_st[:-1])]
    return test, sum(test)
    
missing = [y for x,y in bikeid_gb.apply(count_missing)]

(nrow,ncol)=data.shape
# minimum fraction of missing record
print "="*100
print 'the minimum fraction of missing record is {:.2f}'.format(sum(missing)/float(nrow))
print "="*100
```

## 3. Average amount of time a bike spent at a station

```python
def station_time(df):
    test,_ = count_missing(df)
    starttime = df['starttime'].values
    stoptime = df['stoptime'].values
    staytime = [(x[0]-x[1])/np.timedelta64(1,'s') for ind,x in enumerate(zip(starttime[1:],stoptime[:-1])) if test[ind]==False]
    return np.mean(staytime)

stay_time = bikeid_gb.apply(station_time)

warnings.simplefilter("ignore", RuntimeWarning)

# average amout of time spent at a station
print "="*100
print 'the average amount of time a bike spent at a station is {:.0f} seconds'.format(stay_time.mean())
print "="*100
```

## 4. Exploratory: How the stations are connected

Make an ajacency matrix with row indexed by start station and column by end station, the number of rides gives the weight of the edges. There are 330 stations in total, yet one station is listed only as start station but not as end, so one column is added for that station with nan or zeros.

```python
corr = data.starttime.groupby([data['start station id'],data['end station id']]).count().unstack()
# there is one station only listed as start but not as end, need to add it with zeros to make a square matrix
missed = set(data['start station id']).difference(set(data['end station id'])).pop()
corr[missed]=np.nan
# reorder the columns
corr = corr.reindex_axis(sorted(corr.columns),axis =1)
corr = corr.fillna(0)
```

reset row and column index of the ajacency matrix but keep track of correlation between the ordinal and the original station name using a dict

```python
corr.reset_index(inplace=True,drop = True)
oldnames = list(corr.columns.values)
newnames = [str(x) for x in range(330)]
namedict = dict(zip(oldnames,newnames))
corr.rename(columns=namedict,inplace = True)
```

generate the adjacency matrix plot on log scale with zeros (originally nan values) encoded with white color

```python
def plot_corr(data):
    if type(data) != np.ndarray:
        tmp = data.as_matrix()
    else:
        tmp = data
    lgtmp = np.log(tmp)
    pallete = plt.cm.cool
    pallete.set_bad('w',1.0)
    lgtmp = np.ma.array(lgtmp,mask = np.isinf(lgtmp))
    plt.figure(figsize = (10,10))
    plt.imshow(lgtmp,cmap = pallete, interpolation = 'nearest')
    plt.colorbar()

plot_corr(corr)
```

## 5. Find the communities

It would be interesting to see if the stations form clusters and whether their daily/hourly usage differentiate from one cluster to another. To do that, a network community finding algorithm needs to be used, which can be found in the graph library. 

First, we need to convert corr from a dataframe to a graph adjacency list. To make the algorithm more efficient, we simplify the graph structure of the current system, we assume:
        
        1. weighted but undirected, the bigger weight is used to symmetrize the adjacency matrix
        2. no self-loop, simple graph

```python
test = corr.as_matrix()
cpm = test > test.T
symt = test*cpm + test.T*(~cpm) - np.diag(test.diagonal())
upper = np.triu(symt,k=0)
corr1 = pd.DataFrame(data = upper, columns = newnames, index = newnames)
corr1 = corr1.stack()
corr1.index.set_names(['head','tail'],inplace = True)
corr1 = corr1[corr1!=0]
# corr1.to_csv('adjlist3.txt',sep = ' ')
```

define a function to simultaneously permute the rows and columns of the adjacency matrix. This will be called later when certain permutation is generated by the communty-finding algorithm.

```python
def permutateIndexes(array, perm):
    return array[np.ix_(perm, perm)]
```

define a function <b>commdetect</b>, arguments: a graph ajacency list (e.g., corr1), the chosen algorithm, the number of targeted clusters, 

returns: <b>clusters</b>, which is a list of subgraphs, where each subgraph is a list of nodes

print: the modularity of the final clustering

plot: call plot_corr to plot the adjacency matrix after permutation according to the clustering

```python
gtest = Graph.Read_Ncol('adjlist3.txt',names = True, weights = True, directed = False)

def commdetect(graph,algorithm,n=None):
    """
    commdetect(graph,algorithm,ncluster)
    
    given an undirected but weighted graph; 
    specified community detection algorithm chosen from:
            fastgreedy
            walktrap
            leading_eigenvector
            infomap
            spinglass
    target number of clusters
    
    return comms: a list of subgraphs(clusters) where each element is a subgraph list of nodes
    """
    if algorithm == 'fastgreedy':
        vd = graph.community_fastgreedy(weights = 'weight')
    elif algorithm == 'walktrap':
        vd = graph.community_walktrap(weights = 'weight')
    elif algorithm == 'leading_eigenvector':
        cluster = graph.community_leading_eigenvector(weights = 'weight')
    elif algorithm == 'infomap':
        cluster = graph.community_infomap(edge_weights = 'weight')
    elif algorithm == 'spinglass':
        cluster = graph.community_spinglass(weights = 'weight')
    else:
        raise Exception('no such algorithm')
        return
    
    if (algorithm == 'infomap')|(algorithm == 'leading_eigenvector')|(algorithm == 'spinglass'):
        pass
    else:
        cluster = vd.as_clustering(n)
    print '='*100
    print 'the modularity of the current clustering is {:.3f}'.format(cluster.modularity)
    print '='*100
    subgs = cluster.subgraphs()
    comms = [x.vs['name'] for x in subgs]
    perm = [int(item) for sublist in comms for item in sublist]
    blocked = permutateIndexes(test,perm)
    plot_corr(blocked)
    return comms
    
clusters = commdetect(gtest,'fastgreedy')
```

The three clusters produced by commdetect are labeled in the following analysis as 0, 1, 2; start from the left upper corner to the right-bottom corner. It can be seen that the 0 and 1 clusters are not well separated as it will be further discussed and verified later, while cluster 2 is well separated from the other two. The modularity score also indicates the imperfect clustering.

convert clusters to saparate node lists with original node name, stored in "nodes"

```python
namedict2 = {}
for key,val in namedict.items():
    namedict2[val] = key

nodes = {}
for i in xrange(len(clusters)):
    nodes['cluster'+str(i)] = [namedict2[key] for key in clusters[i]]
```

Add two factor features to data, indicating the clusters [0 1 2] that the start/end station belong to

```python
data['start_comm'] = np.zeros(data.shape[0])
data['end_comm']= np.zeros(data.shape[0])

data.start_comm[data['start station id'].isin(nodes['cluster2'])] = 2
data.start_comm[data['start station id'].isin(nodes['cluster0'])] = 0
data.start_comm[data['start station id'].isin(nodes['cluster1'])] = 1

data.end_comm[data['end station id'].isin(nodes['cluster0'])] = 0
data.end_comm[data['end station id'].isin(nodes['cluster1'])] = 1
data.end_comm[data['end station id'].isin(nodes['cluster2'])] = 2

data.start_comm = data.start_comm.astype(int)
data.end_comm = data.end_comm.astype(int)
```

## 6. Daily and Hourly usage patterns 

The middle time (meantime = (starttime + stoptime)/2) is used to define the time of the ride, which is used in the time series analysis

### investigate rides per day for different types of rides, weekday vs weekend & holidays

generate a new dataframe ts_data to count number of rides of each type, ts_data, where <b>comm</b> is a column indicating the start and end community as <b>(start,end)</b>

```python
meantime = data.starttime + (data.stoptime-data.starttime)/2
ts_data = pd.DataFrame({'comm':zip(data.start_comm,data.end_comm),'ridecount':np.ones(data.shape[0])},index = meantime)
ts_data.index.name = 'ridetime'
ts_data.head()
```

define a resample function to resample the data by day

```python
def resample_rd(df,freq = 'H'):
    return df.resample(freq,how = 'sum')
#daily resampled
tsrd = ts_data.groupby('comm').apply(resample_rd,freq = 'D')
tsrd.index.names = [None,None]
tsrd = tsrd.unstack(level = 0)
tsrd.columns = tsrd.columns.droplevel()
# remove the last day (11/26/2013) which has only ending but no start
tsrd = tsrd[:-1]
```

define a function <b> endholiplot </b> to detect weekends and three holidays during the two months, they are Columbus day (10/14), Halloween(10/31) and Veterans Day (11/11)

```python
def endholiplot(axis):
    axis.legend(loc='best',frameon=False)
    for ind,date in enumerate(list(tsrd.index)):
        if date.weekday() == 5:    # detect if it is a Saturday
            axis.axvspan(tsrd.index[ind],tsrd.index[ind+1],color=(0,0,0.5),alpha = 0.3)
# interesting finding here, (0,0) always plummet on weekends, while this is not true for (1,1) and (2,2)
        elif (date == pd.to_datetime('2013-10-14'))|(date == pd.to_datetime('2013-10-31'))|(date == pd.to_datetime('2013-11-11')):
            axis.axvline(tsrd.index[ind],color = (0.5,0,0.5),alpha = 0.7,linewidth = 2)
```

Compare the intra community ride types

```python
fig = plt.figure(figsize = (6,5))
tsrd[[(0,0),(2,2),(1,1)]].plot(legend = False,color = ['r','y','b'])
ax = plt.gca()
ax.set_title('daily number of rides')
endholiplot(ax)
remove_border()
```

Two main observations:

1. the daily usage in (0,0) (red) always plummet around the weekends, it is not generally true for intra-community rides in (1,1) and (2,2), even the opposite trend is seen, namely, the usage peaks on weekends, especially (2,2) (yellow). This indicates that cluster 0 might be more of a business district, bikes are more used as a commute to and from work whereas clusters 1 and 2 can be housing neigborhood or mixed with business district. This postulation is buttressed again by the following hourly usage plot.

2. among the three holidays, it seems only the Halloween has a dramatic influence on the usage, creating a dip during the weekdays of that week. 

Compare daily usage of inter-community rides

```python
fig, axes=plt.subplots(figsize = (18,5),nrows=1, ncols=3)
cts = range(3)
cts.append(0)
for i in xrange(3):
    axis = axes[i]
    tsrd[[(cts[i],cts[i+1]),(cts[i+1],cts[i])]].plot(ax = axis,color = ['r','b'])
    endholiplot(axis)
fig.tight_layout()
```

Observations:

    1. the total number of records are very different, a lot more of records btw (0,1), the ones between 0 and 2 are the fewest.
    2. the daily usage between communities 0 and 1 is not very high on weekends, a few are the lowest in that week. In contrast, the usage between communities 1 and 2 always peaks on weekends, indicating non-work activity use of the bike. To the opposite, the usage between communities 0 and 2 always plummet on weekends. 
    3. very interestingly, a sharp difference is shown on Halloween, where the first two plots show consistent usage drop in both direction while the last one shown drop in the (2,0) direction, nobody is going to 0, probably workplace, not for celebration. Yet, (0,2) peaks! People went from work to 2 for celebration ?!
    
    To summarize, this is again a strong indication that 0 is likely to be close to workplace, business district, 1 is maybe mixed, 2 is housing neighborhood, or some place for fun

### Investigate hourly bike usage among different ride types

extract the hour from the meantime timestamp, make a dataframe, <b>tsrh</b>, which shows how many rides for each ride type hourly, from 0 to 23rd, sum up all records from all the days.

```python
hourind = [x.hour for x in meantime]
tsrh = ts_data
tsrh['hour'] = hourind

tsrh = tsrh.groupby(['comm','hour']).count('ridecount')
tsrh = tsrh.unstack(level = 0)
tsrh.columns = tsrh.columns.droplevel()
tsrh.reset_index(inplace = True)
tsrh.columns.names = [None]
tsrh.head()
```

```python
xticks = range(0,24,4)
fig, axes=plt.subplots(figsize = (15,10),nrows=3, ncols=3)
for i in xrange(3):
    for j in xrange(3):
        axis = axes[i][j]
        tsrh.plot(kind = 'bar', x ='hour',y = (i,j),ax = axis,label = '({},{})'.format(i,j),xticks = xticks)
        axis.set_xticklabels(list(range(0,24,4)),rotation = 0)
        axis.legend(frameon = False)
        axis.set_xlabel('')
fig.tight_layout()
```

Observations and thoughts:
    1. Generally, the hourly usage peaks at rush hours, e.g., 8 am and 5-6 pm, take a look at (0,0),(1,0) for example.
    
    2. However, take (0,2) and (2,0), one can only spot single peak area, for (0,2) is around 6 pm while for (2,0) is around 8 am. This is a strong evidence that the bike tranffic between 0 and 2 are largely <b>unilateral</b> at the two times of a day. Presumably, people go to work (2 to 0) in the morning, and get back home in the afternoon (0 to 2). The opposite flow is much smaller. Similar but weaker pattern also shows in (2,1) and (1,2), indicating 1 is more of a workplace than 2
    
    3. The hourly usage should be very different between weekdays and weekends or holidays, which should give more insight. due to the limit of time, the analysis ends here.
